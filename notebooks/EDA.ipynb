# Titanic EDA\nExploratory analysis for missing data, feature distributions, and survival patterns.

```python
from pathlib import Path\nimport sys\nimport pandas as pd\nimport seaborn as sns\nimport matplotlib.pyplot as plt\nROOT = Path.cwd().resolve().parents[0] if Path.cwd().name == 'notebooks' else Path.cwd()\nsys.path.append(str(ROOT / 'src'))\nfrom preprocess import load_raw_data, add_engineered_features\nsns.set_theme(style='whitegrid')
```

```python
df = load_raw_data(ROOT / 'data' / 'raw' / 'train.csv')\ndf = add_engineered_features(df)\ndf.head()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 4))\nsns.histplot(df['Age'], kde=True, ax=axes[0])\naxes[0].set_title('Age Distribution')\nsns.histplot(df['Fare'], kde=True, ax=axes[1])\naxes[1].set_title('Fare Distribution')\nsns.countplot(data=df, x='Pclass', ax=axes[2])\naxes[2].set_title('Passenger Class Distribution')\nplt.tight_layout()
```

```python
corr_df = df[['Survived', 'Pclass', 'Age', 'Fare', 'SibSp', 'Parch', 'FamilySize', 'IsAlone']].copy()\nplt.figure(figsize=(9, 6))\nsns.heatmap(corr_df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')\nplt.title('Correlation Heatmap')\nplt.tight_layout()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 4))\nsns.barplot(data=df, x='Sex', y='Survived', ax=axes[0])\naxes[0].set_title('Survival by Sex')\nsns.barplot(data=df, x='Pclass', y='Survived', ax=axes[1])\naxes[1].set_title('Survival by Pclass')\nsns.boxplot(data=df, x='Survived', y='Age', ax=axes[2])\naxes[2].set_title('Age vs Survival')\nplt.tight_layout()
```
